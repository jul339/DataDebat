```python
import json
import re
from pathlib import Path
import pandas as pd

def json_to_shifts_df(json_path: Path) -> pd.DataFrame:
    with open(json_path, encoding="utf-8") as f:
        results = json.load(f)
    rows = []
    for word, info in results.get("words", {}).items():
        shifts = info.get("temporal_shifts", {})
        for k, v in shifts.items():
            m = re.search(r"(\d{4}).*?(\d{4})", str(k))
            if not m:
                continue
            y1 = int(m.group(1))
            y2 = int(m.group(2))
            rows.append({"word": word, "year1": y1, "year2": y2, "distance": float(v)})
    df = pd.DataFrame(rows)
    if df.empty:
        return df
    df["mid_year"] = (df["year1"] + df["year2"]) / 2
    return df.sort_values(["word", "year1"]).reset_index(drop=True)

df_shifts = json_to_shifts_df(Path("DataDebat/results/overton_analysis.json"))
```
